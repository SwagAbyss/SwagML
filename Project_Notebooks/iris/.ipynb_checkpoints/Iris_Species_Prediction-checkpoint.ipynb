### __Violinplot__
__Violin plots have many of the same summary statistics as box plots:__

- the white dot represents the median
- the thick black bar in the center represents the interquartile range
- the thin gray line represents the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the interquartile range.

for reference : https://www.labxchange.org/library/items/lb:LabXchange:46f64d7a:html:1

```python
# creates a figure and multiple subplots using the matplotlib library
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,5), dpi=200)

# flattens the 2D array of subplots into a 1D array
axes =axes.flat

# creates a list of column names that will be used to plot the scatterplot
numfeatures = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width']
target = 'Class'

# for loop that will iterate through the list of column names
for index, column in enumerate(numfeatures):
    sns.scatterplot(
        data=data,
        x=column,
        y=target,
        linewidths=0.25,
        color=(list(plt.rcParams['axes.prop_cycle'])*2)[index]['color'],
        ax=axes[index]
)
    
    # sets the title of the current subplot to the current column name
    axes[index].set_title(f'{column} vs {target}', fontsize=12)
    axes[index].set_xlabel(column, size=12)
    axes[index].set_ylabel(target, size=12)
    axes[index].grid()
    
    
# set the title for the plot
plt.suptitle('Target vs Numerical Features', size=20)

# adjusts the spacing between subplots to optimize the layout
plt.tight_layout()

# displays the figure on the screen
plt.show()
```

```python
# Select the features
features = data[['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width', 'Class']]
```

## __5.4 Scaling the data__

```python
from sklearn.preprocessing import StandardScaler

# creating an object for the class
scale = StandardScaler()

# scaling the data 
scaled_features = scale.fit_transform(features)
```

# __6. Building a Model__

## __6.1 Determine Features & Target Variables__

```python
# Splitting the dataset into target(X) and features(y)
X = scaled_features
y = data.Class
```

## __6.2 Split the dataset into Traing and Test sets__

```python
# Splitting the dataset into train and test
from sklearn.model_selection import train_test_split

Xtrain, Xtest, ytrain, ytest = train_test_split(
    X, 
    y, 
    random_state=2, 
    test_size=0.33
)
```

```python
# checking the row and columns of the train and test data
print('xtrain shape: ', Xtrain.shape)
print('xtest shape: ', Xtest.shape)
print('ytrain shape: ', ytrain.shape)
print('ytest shape: ',ytest.shape)
```

## __6.3 Train and Evaluate the Model__

### __Finding the best parameters in the SVC Algorithm__

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# setting the parameters
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma' : [0.1, 0.01, 0.001],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid']
}

# create a grid search object
grid_search = GridSearchCV(SVC(), param_grid, verbose=1)

# fit the grid search to the data
grid_search.fit(Xtrain, ytrain)

# finding the best parameters
grid_search.best_params_
```

## __6.4 Training the model with best parameters__

```python
# re-create a SVC model with the best parameters
svc = grid_search.best_estimator_

# fit the model to the training data
svc.fit(Xtrain, ytrain)

# predict the labels for the test data
ypred = svc.predict(Xtest)

# checking bias-variance tradeoff
train_accuracy = svc.score(Xtrain, ytrain)
test_accuracy = svc.score(Xtest, ytest)

# converting accuracy score into dataframe
dataframe = pd.DataFrame(
    [train_accuracy*100, test_accuracy*100], 
    index=['Train Accuracy', 'Test Accuracy'], 
    columns=['Value']
)
print(dataframe)

# classification report for the model
report = classification_report(ytest, ypred)
print(report)

# using heatmap to plot confusion matrix graph
sns.heatmap(confusion_matrix(ytest, ypred), annot=True)

print('The accuracy of the SVC is',accuracy_score(ytest,ypred),'\n')
```

## __6.5 Train the model using ensemble model algorithms__

```python
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier, RandomForestClassifier, ExtraTreesClassifier

# Writing a function for using different types of algorithms
def select_model_predict(model):
    
    # training the model
    model.fit(Xtrain, ytrain)
    global ypred
    
    # predict the labels for the test data
    ypred = model.predict(Xtest)
    
    # checking Bias-variance tradeoff
    train_accuracy = model.score(Xtrain, ytrain)
    test_accuracy = model.score(Xtest, ytest)
    
    # converting Accuracy score into dataframe
    dataframe = pd.DataFrame(
        [train_accuracy*100, test_accuracy*100], 
        index=['Train Accuracy', 'Test Accuracy'], 
        columns=['Value']
    )
    
    print(model, '\n')
    
    print(dataframe, '\n')
    
    # using heatmap to plot confusion matrix graph
    sns.heatmap(confusion_matrix(ytest, ypred), annot=True)
    
    # classification report for the model
    report = classification_report(ytest, ypred)
    print(report)
    
    return model
```

### __Using AdaBoost Classifier Algorithm__

```python
ada = select_model_predict(AdaBoostClassifier())
```

### __Using Gradient Boosting Classifier Algorithm__

```python
gbc = select_model_predict(GradientBoostingClassifier())
```

### __Using Extreme Gradient Boosting Classifier Algorithm__

```python
from xgboost import XGBClassifier

xgbc = select_model_predict(XGBClassifier())
```
