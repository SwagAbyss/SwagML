```python
# H0 : it is dynamic data
# H1 : it is static data

from statsmodels.tsa.stattools import adfuller

def adfuller_check(column):
    pvalue=adfuller(column)[1]
    print(pvalue)
    if pvalue <=0.05:
        print('it is static data')
    else:
        print('it is dynamic data')
    
```

```python
adfuller_check(data['Passengers'])
```

## __Changing the data to Static data using Difference method__

```python
data['year_static'] = data['Passengers'] - data['Passengers'].shift(12)
data['year_static'].plot(ylabel='Passengers')
```

```python
for i in range(1, 13):
    print(i)
    adfuller_check((data['Passengers'] - data['Passengers'].shift(i)).dropna())
```
