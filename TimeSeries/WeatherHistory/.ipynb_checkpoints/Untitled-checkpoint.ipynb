# __Data Preperation__

```python
data = raw_data.copy(deep=True)
```

```python
data.columns
```

```python
data.columns = data.columns.str.replace(' ', '_')
```

```python
data = data.drop(['Loud_Cover', 'Daily_Summary'], axis=1)
```

```python
data['Formatted_Date'] = pd.to_datetime(data['Formatted_Date'], format='%Y-%m-%d', utc=True)
```

```python
data.set_index('Formatted_Date', inplace=True)
```

# __Exploratory Data Analysis__

```python
data['Precip_Type'].fillna('rain', inplace=True)
```

## __Univariate Analysis__

```python
data.nunique()
```

### __Target Distribution__

```python
weather_summary = data['Summary'].value_counts().reset_index()
weather_summary.columns = ['Weather', 'Counts']
weather_summary
```

### __Continuos Numerical Feature Distribution__

```python
numcols = data.select_dtypes(include=np.number).columns
```

```python
fig, axes = plt.subplots(ncols=4, nrows=2, figsize=(15, 6))

axes = axes.flat

for index, column in enumerate(numcols):
    
    sns.histplot(
        data=data,
        x=column,
        kde=True,
        ax=axes[index]
    )

plt.tight_layout()
```

## __Bivariate Analysis__

```python
fig, axes = plt.subplots(ncols=2, nrows=2, figsize=(15, 12))
axes = axes.flat
number_cols = ['Temperature_(C)', 'Humidity', 'Apparent_Temperature_(C)', 'Wind_Speed_(km/h)']
for index, column in enumerate(number_cols):
    sns.barplot(
        data=data, 
        x='Summary', 
        y=column, 
        hue='Precip_Type', 
        palette='cubehelix',
        ax=axes[index]

    )
    plt.setp(axes[index].get_xticklabels(), rotation=90)
plt.title('Temperature vs Summary')

plt.tight_layout(pad=2)
plt.show()
```
